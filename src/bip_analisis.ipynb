```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_bip['comuna'].value_counts().plot(kind='barh')
```

```python
df_bip['linea'].value_counts().plot(kind='barh')
```

```python
sns.catplot(data=df_bip, kind='count', x='linea')
```

```python
plt.figure()
sns.catplot(data=df_bip, kind='count', y='comuna',
            order=df_bip['comuna'].value_counts(ascending=True).index,
            height=10, aspect=1, palette="flare")
plt.title('Cantidad de Centros Bip! por Comuna', fontsize=22)
plt.xlabel('Cantidad', fontsize=18)
plt.ylabel('Comunas', fontsize=18)
plt.show()
```

```python
plt.figure()
sns.catplot(data=df_bip, kind='count', y='linea',
            order=df_bip['linea'].value_counts(ascending=True).index,
            height=10, aspect=1, palette="flare")
plt.title('Cantidad de Centros Bip! por Línea', fontsize=22)
plt.xlabel('Cantidad', fontsize=18)
plt.ylabel('Línea', fontsize=18)
plt.show()
```

```python
df_comunas = pd.DataFrame(df_bip['comuna'].value_counts().reset_index())
df_comunas.rename(columns={'index':'comuna', 'comuna':'cantidad'}, inplace=True)
df_comunas
```

```python
df_top_comunas = df_comunas[df_comunas['cantidad'] >= 9]
df_top_comunas
```

```python
plt.figure(figsize=(10,10))
plt.pie(data=df_top_comunas, x=df_top_comunas['cantidad'], labels=df_top_comunas['comuna'], 
        textprops={'size':16} ,autopct='%1.1f%%', colors=sns.color_palette('Set2'))
plt.title(
    label="Top 5 Comunas con más Centros Bip!", 
    fontdict={"fontsize":22},
    pad=2
)
plt.show()
```
